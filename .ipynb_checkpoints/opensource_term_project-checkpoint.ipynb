```python
# 큰 화면 찾기

for i in numpy.arange(0.51,0.99, 0.02):
    p_list = pag.locateAllOnScreen("C:/Users/tlsrl/opensource_term/window1.png", confidence=i)
    p_list = list(p_list)
    if len(p_list) == 1:
        print(i)
        print(p_list)
        break
```

```python
# 컴퓨터에서 연산하기 위한 행렬 생성
pixel = p_list[0][2]// 18
col = p_list[0][2] // pixel
row = (p_list[0][3] - pixel*2) // pixel
matrix = numpy.zeros((row, col))
print(matrix)

# 좌표 중앙 리스트
ctr_list = []
for i in range(row):
    ctr_row = []
    for k in range(col):
        ctr_row.append([p_list[0][0] + pixel//2 + k*pixel, p_list[0][1]+ pixel*2 + pixel//2 + i*pixel])
    ctr_list.append(ctr_row)

ctr_list
```

```python
# 타겟 찾기

pic_list = pag.locateAllOnScreen("C:/Users/tlsrl/opensource_term/one1.png", confidence=0.70)
pic_list = list(pic_list)
#print(len(pic_list))
#print(pic_list)
ctr = []

for p in pic_list:
    if not ctr:
        ctr.append(pag.center(p))
        #print(pag.center(p))
    else:
        #print(ctr[-1][0]-pag.center(p)[0])
        #print(ctr[-1][1]-pag.center(p)[1])
        for i in ctr:
            overlapped = False
            if abs(i[0]-pag.center(p)[0])<=10 and abs(i[1]-pag.center(p)[1])<=10:
                overlapped = True
                break
        if not overlapped:
            ctr.append(pag.center(p)) 
            #print(pag.center(p))
print(len(ctr))

# 타겟으로 자동 마우스 이동

'''
for i in ctr:
    pag.moveTo(i)
    time.sleep(0.5)


for i in range(len(matrix)):
    for j in range(len(matrix[i])):
        print(i, j)
'''

for i in ctr:
    a = (i[0] - p_list[0][0]) // pixel # col
    b = (i[1] - p_list[0][1] - pixel*2) // pixel # row
    matrix[b][a] = 1
```

```python
# 클릭 못하는 곳 지정
wall1 = Image.open("C:/Users/tlsrl/opensource_term/wall1.png")
wall2 = Image.open("C:/Users/tlsrl/opensource_term/wall2.png")
#print(wall1.size)
pix1 = numpy.array(wall1)
pix2 = numpy.array(wall2)
#print(pix[14][14])
for r in range(row):
    for c in range(col):
        if matrix[r][c]==0:
            screen = ImageGrab.grab()
            color_pos = x, y = int(ctr_list[r][c][0]), int(ctr_list[r][c][1])
            color = screen.getpixel(color_pos)
            if color[0]==pix1[14][14][0] and color[1]==pix1[14][14][1] and color[2]==pix1[14][14][2]:
                matrix[r][c] = 99
            elif color[0]==pix2[14][14][0] and color[1]==pix2[14][14][1] and color[2]==pix2[14][14][2]:
                matrix[r][c] = 99
```

```python
print(matrix)
print(matrix.shape)
answer_list = numpy.zeros((row, col))
print(answer_list.shape)
for r in range(row):
    for c in range(col):
        if matrix[r][c]!= 0 and matrix[r][c]!= 9:
            if 0<=r-1<=row and 0<=c-1<=col and matrix[r-1][c-1]==0:
                answer_list[r-1][c-1] +=1
            if 0<=r-1<=row and 0<=c<=col and matrix[r-1][c]==0:
                answer_list[r-1][c] +=1
            if 0<=r-1<=row and 0<=c+1<=col and matrix[r-1][c+1]==0:
                answer_list[r-1][c+1] +=1
            if 0<=r<=row and 0<=c-1<=col and matrix[r][c-1]==0:
                answer_list[r][c-1] +=1
            if 0<=r<=row and 0<=c+1<=col and matrix[r][c+1]==0:
                answer_list[r][c+1] +=1
            if 0<=r+1<=row and 0<=c-1<=col and matrix[r+1][c-1]==0:
                answer_list[r+1][c-1] +=1
            if 0<=r+1<=row and 0<=c<=col and matrix[r+1][c]==0:
                answer_list[r+1][c] +=1
            if 0<=r+1<=row and 0<=c+1<=col and matrix[r+1][c+1]==0:
                answer_list[r+1][c+1] +=1
print(answer_list)
```

```python
answer_r = numpy.argmax(answer_list) // col
answer_c = numpy.argmax(answer_list) % col
pag.moveTo(ctr_list[answer_r][answer_c][0], ctr_list[answer_r][answer_c][1], 1)
```
