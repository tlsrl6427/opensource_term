```python
from PyKomoran import Komoran, DEFAULT_MODEL

komoran = Komoran(DEFAULT_MODEL['LIGHT'])
```

```python
def pos(pos):
    dic = {
        'EC': switch_EC(pos),
        'EF': switch_EF(pos),
        'JX': switch_JX(pos),
        'ETM': switch_ETM(pos),
        'JKB': switch_JKB(pos),
        'NP': switch_NP(pos),
    }
    
    return dic.get(pos)
    
```

```python
and_change = False

def switch_word(key):
    dic = {
        "오늘" : "금일",
        "너" : "그대",
        "왜" : "어찌하여",
        "뭐" : "무얼",
        "저" : "소인",
        
    }
    
    changed_word = dic.get(key, key)
    return changed_word


def switch_EC(key):
    dic = {
        "습니다": "사옵니다",
        "ㅂ니다": "옵니다",
        "ㄴ다": "오",
        "다": "소",
        #"아서": "서",
        #"아도": "도",
        "자": "지",
        "어": "오",
        "에요": "오",
        "어요": "오",
    
    }
    changed_word = dic.get(key, key)
    return changed_word


def switch_EF(key):
    dic = {
        "어": "시오",
        "어요": "오",
        "죠": "오",
        "잖아": "다",
        "ㅂ니까": "오",
        "나요" : "시오",
        "습니다": "사옵니다",
        "냐": "오",
        "을까": "겠느냐",
        "아": "느냐",
        "세요": "시오",
        "니": "는가",
    }
    changed_word = dic.get(key, key)
    return changed_word

def switch_JX(key):
    dic = {
        "ㄴ": "는",
        "이야": "이다",
    }
    changed_word = dic.get(key, key)
    return changed_word

def switch_ETM(key):
    dic = {
        "ㄴ": "는",
        
    }
    changed_word = dic.get(key, key)
    return changed_word

def switch_JKB(key):
    dic = {
        "한테": "에게"
    }
    changed_word = dic.get(key, key)
    return changed_word

def switch_NP(key):
    dic = {
        "ㄴ": "는",
        "나": "소인",
        "저": "소인",
        "너": "그대",
        "뭐": "무얼",
    }
    changed_word = dic.get(key, key)
    return changed_word

def switch_IC(key):
    dic = {
        "그래": "그리하오",
    }
    changed_word = dic.get(key, key)
    return changed_word
```

```python
text = "정숙하세요"
print("Input sentense: "+text)
a = komoran.get_token_list(text)
print(a)
text2=""
for i in range(len(a)):
    text1 = ""
    previous_text = ""
    print(a[i].get_morph())
    print(a[i].get_pos())
    print(a[i].get_begin_index())
    print(a[i].get_end_index())
    for j in range(a[i].get_begin_index(), a[i].get_end_index()):
        text1 += text[j]
    print(text1)
    for k in range(a[i-1].get_begin_index(), a[i-1].get_end_index()):
         previous_text += text[k]
    print("\n")
    if(a[i-1].get_end_index() == a[i].get_begin_index()-1):
        text2 += " "
    
    if(previous_text != text1):
        if (a[i].get_pos() == 'EC'):
            text2 += text1.replace(text1, switch_EC(a[i].get_morph()))
        elif (a[i].get_pos() == 'EF'):
            text2 += text1.replace(text1, switch_EF(a[i].get_morph()))
        elif (a[i].get_pos() == 'JX'):
            if (a[i-1].get_morph() == '나'):
                text2 += text1.replace("는", "은")
            else:
                text2 += text1.replace(text1, switch_JX(a[i].get_morph()))
        elif (a[i].get_pos() == 'ETM'):
            text2 += text1.replace(text1, switch_ETM(a[i].get_morph()))
        elif (a[i].get_pos() == 'NP'):
            text2 += text1.replace(text1, switch_NP(a[i].get_morph()))
        elif (a[i].get_pos() == 'IC'):
            text2 += text1.replace(text1, switch_IC(a[i].get_morph()))
        elif (a[i].get_pos() == 'JKB'):
            if (a[i-1].get_morph() == '저'):
                text2 = text1.replace("와", "과")
            else:
                text2 += text1.replace(text1, switch_JKB(a[i].get_morph()))
        else:
            text2 += text1.replace(text1, switch_word(a[i].get_morph()))
    
    else:
        if (a[i].get_pos() == 'EC'):
            if(switch_EC(a[i].get_morph()) == a[i].get_morph()):
                text2 = text2.replace(a[i-1].get_morph(), text1)
            else:
                text2 += text1.replace(text1, switch_EC(a[i].get_morph()))
            
        elif (a[i].get_pos() == 'EF'):
            text2 += text1.replace(text1, switch_EF(a[i].get_morph()))
        elif (a[i].get_pos() == 'JX'):
            if (a[i-1].get_morph() == '나'):
                text2 += text1.replace("는", "은")
            else:
                text2 += text1.replace(text1, switch_JX(a[i].get_morph()))
        elif (a[i].get_pos() == 'ETM'):
            text2 += text1.replace(text1, switch_ETM(a[i].get_morph()))
        elif (a[i].get_pos() == 'NP'):
            text2 += text1.replace(text1, switch_NP(a[i].get_morph()))
        elif (a[i].get_pos() == 'IC'):
            text2 += text1.replace(text1, switch_IC(a[i].get_morph()))
        elif (a[i].get_pos() == 'JKB'):
            if (a[i-1].get_morph() == '저'):
                text2 = text1.replace("와", "과")
            else:
                text2 += text1.replace(text1, switch_JKB(a[i].get_morph()))
        else:
            text2 = text2.replace(a[i-1].get_morph(), text1)

        
print("Output sentense: "+ text2)
```
